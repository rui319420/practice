```python
import matplotlib.pyplot as plt

x = [10000, 100000, 1000000, 5000000, 10000000]
y_before = [12, 19, 156, 873, 1848]
y_after = [8, 11, 15, 18, 31]

plt.plot(x, y_before, label="before_index")
plt.plot(x, y_after, label="after_index")
plt.xlabel('number_of_data')
plt.ylabel('time(ms)')
plt.title('Graphs of database')
plt.legend()
plt.show()
```
